# 16) How do I handle missing values in pandas?

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.tail()
```

```python
ufo.isnull().tail()
#isnull missing value ları false olarak gösterir
```

```python
ufo.notnull().tail()
#notnull missing value ları true  olarak gösterir
```

```python
ufo.isnull().sum()
#kac tane null deger oldugunu gösterir.
```

```python
pd.Series([True, False, True]).sum()
```

```python
ufo.City.isnull()
#series şeklinde boolean deger döndürür.
```

```python
ufo[ufo.City.isnull()]
#df şeklinde boolean olan degerleri getirir.
```

```python
ufo.shape
#row ve columns sayısı
```

```python
ufo.dropna(how="any").shape
#bir satırda bir tane bile olsun null deger varsa o satırları kaldırır ve kalanın sayısını gösterir.
# yani 2486 tane row kaldırmış

```

```python
ufo.shape
# değişmedi row degeri. cünkü kalıcı değişiklik yapmadık.
```

```python
ufo.dropna(how="all").shape
# aynı değer gelmesinin nedeni row un tamamının null değer oldugu satır bulunmaması.
```

```python
ufo.dropna(subset = [ "City", 'Shape Reported'], how ="all").shape
```

```python
ufo['Shape Reported'].fillna(value="VARIOUS",inplace=True)
```

```python
ufo['Shape Reported'].value_counts()
#dropna default olarak true ve nan degerleri dahil etmiyor
```

```python
ufo['Shape Reported'].value_counts(dropna=False)
#dropna yı true yaparak nan degerleri de getirmiş olduk
```

# 19)How do I select multiple rows and columns from a pandas DataFrame?

```python
import pandas as pd
```

```python
ufo =pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
ufo.loc[0 , :]
#loc[row , columns]
#o ıncı satırın bütün columnslarını getir.
```

```python
ufo.loc[[0,1,2], :]
#0 ,1 ,2 satırlarının bütün columnslarını getir.
```

```python
ufo.loc[0:2, :]
#2 dahil olmali. inclusive deger.
```

```python
ufo.loc[0:2]
#tek bir değer yazarsak sadece row a göre davranır. 0 ,1,2. satırları getir.
```

```python
ufo.loc[:, ["City", "State"]]
#sadece City ve State columnslarını bütün row larıyla getirir.
```

```python
ufo.loc[:, "City": "State"]
# City den State e kadar bütün columnsları getirir.
```

```python
ufo.head(3).drop("Time", axis=1)
```

```python
ufo[ufo.City =="Oakland"]
```

```python
ufo.loc[ufo.City =="Oakland", :]
```

```python
ufo.loc[ufo.City =="Oakland", "State"]
#ufo.loc[ufo.City =="Oakland", :].State.. bu kısım zincirleme index diye bilinir
# ama bazen bazı problemlere neden olabilir.
# o yuzden üstteki daha doğru bir seçimdir.
```

```python
ufo.columns
```

```python
#integer position şeklinde kullanılır
ufo.iloc[:,0:4]
#0 dan 4 e kadar (4 dahil degil yani exclusive) bütün rowları getirir.
```

```python
ufo.iloc[0:3, :]
```

```python
ufo[0:2]
#ufo.iloc[0:2,:]
```
