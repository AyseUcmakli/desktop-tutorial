# BARPLOT

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
company = ["GOOGL","AMZN","MSFT", "FB"]
revenue = [90,136,89,27]
profit =[40,2,34,12]
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company)

plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos, revenue, label ="revenue")
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos, revenue, label ="revenue")
plt.bar(ypos, profit, label= "profit")
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos, company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2, revenue, label ="revenue")
plt.bar(xpos+0.2, profit, label= "profit")
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.xlabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2, revenue,width=0.4, label ="revenue")
plt.bar(xpos+0.2, profit,width=0.4, label= "profit")
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.title("US Tech Stocks")

plt.barh(xpos-0.2, revenue, label ="revenue")
plt.barh(xpos+0.2, profit, label= "profit")
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.title("US Tech Stocks")

plt.barh(xpos-0.2, revenue, label ="revenue")
plt.barh(xpos+0.2, profit, label= "profit")
plt.legend()
```

# PE CHART

```python
exp_vals = [1400, 600,300, 410, 250]
exp_labels = ["home rent", "food", "phone/nternet bill", "car", "other utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels =exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels =exp_labels, radius =1.5)#radius yarcap
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels =exp_labels, radius =1.5, autopct="%0.2f%%")#radius yarcap
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels =exp_labels, radius =1.5, autopct="%0.2f%%",shadow = True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels =exp_labels, radius =1.5, autopct="%0.2f%%",shadow = True, explode=[0,0.5,0,0,0,], startangle =45)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels =exp_labels, radius =1.5, autopct="%0.2f%%",shadow = True, explode=[0,0.1,0.1,0,0,], startangle =180)
plt.show()
```
